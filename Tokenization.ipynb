What is Tokenization?
Tokenization is the process of breaking down text into smaller units called tokens. These tokens could be words, sub-words, or characters. Since AI models don't understand raw text, they need these tokens converted into numbers (specifically, integers) to process them. The quality of this step directly impacts the performance of any downstream AI task, from chatbots to autonomous agents

This is the foundation. You take a string, find all the unique characters, and map them to numbers.

```python
# The text we want to encode
text = "Hello Agentic AI"
# Create a sorted list of unique characters
chars = sorted(list(set(text)))
# Create the mapping (String to Integer)
stoi = {ch:i for i,ch in enumerate(chars)}
# Encode the string into integers
encoded = [stoi[c] for c in text]
print(encoded)
# Output: [34, 25, 44, 44, 47, 0, 26, 27, 35, 42, 48, 39, 13, 0, 26, 29]
```

Using spaCy (For Linguistic Analysis)
If you are interested in how Agents understand grammar (nouns, verbs, etc.), spaCy is the tool.

What you will learn: Tokenization combined with Part-of-Speech (POS) tagging.

How to run: In a new Colab cell, install the model and run the code below

```python
# In Colab, run this in a cell
# !pip install spacy
# !python -m spacy download en_core_web_sm

import spacy

# Load the model
nlp = spacy.load("en_core_web_sm")
# Process the text
doc = nlp("The autonomous agent needs to tokenize this sentence.")
# Print the tokens
for token in doc:
    print(f"Token: {token.text} \t POS: {token.pos_}")
```

```python
# Install the tiktoken library for accurate token counting
# !pip install tiktoken

import tiktoken

def count_tokens(text, model="cl100k_base"):
    enc = tiktoken.get_encoding(model)
    tokens = enc.encode(text)
    return len(tokens)

# --- Interactive loop ---
print("Token Counter (type 'quit' to exit)\n")

while True:
    text = input("Enter your text: ")
    if text.lower() == "quit":
        break

    token_count = count_tokens(text)
    char_count = len(text)
    word_count = len(text.split())

    print(f"  Tokens : {token_count}")
    print(f"  Words  : {word_count}")
    print(f"  Chars  : {char_count}\n")
```
