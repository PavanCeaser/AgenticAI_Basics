Tokenization gives you numbers, but embeddings give you meaning. An embedding is a dense vector (list of numbers) that represents the semantic meaning of a token. Words with similar meanings have similar embeddings.

### Cell 2: Download and Load Embedding Model
```python
# Cell 2: Load pre-trained embeddings (first time only)
import gensim.downloader as api
import numpy as np
import warnings
warnings.filterwarnings('ignore')

print("🔄 Loading embedding model (this downloads ~1.6GB once)...")
print("⏱️ This will take 2-3 minutes on first run\n")

# Load the model
model = api.load("glove-wiki-gigaword-50")

print("\n✅ Model loaded successfully!")
print(f"📊 Vocabulary size: {len(model.key_to_index)} words")
print(f"📐 Embedding dimension: {model.vector_size}")
```

### Cell 3: Explore Individual Word Vectors
```python
# Cell 3: Look inside a word embedding
import numpy as np

# Get vector for "king"
king_vector = model['king']

print("=" * 50)
print("🔍 EXAMINING A WORD EMBEDDING")
print("=" * 50)
print(f"\nWord: 'king'")
print(f"Vector shape: {king_vector.shape}")
print(f"First 10 numbers: {king_vector[:10]}")
print(f"Vector magnitude: {np.linalg.norm(king_vector):.3f}")
print(f"Min value: {king_vector.min():.3f}")
print(f"Max value: {king_vector.max():.3f}")

# Compare with another word
queen_vector = model['queen']
print(f"\n'queen' first 10 numbers: {queen_vector[:10]}")
print(f"Are they similar? First 3 dimensions: {np.allclose(king_vector[:3], queen_vector[:3], atol=0.1)}")
```

### Cell 4: Find Similar Words (Semantic Search)
```python
# Cell 4: Find words with similar meanings
print("=" * 50)
print("🎯 FINDING SIMILAR WORDS")
print("=" * 50)

words_to_test = ['king', 'computer', 'happy', 'running']

for word in words_to_test:
    print(f"\n📌 Words most similar to '{word}':")
    similar_words = model.most_similar(word, topn=5)
    for similar_word, score in similar_words:
        print(f"   {similar_word}: {score:.3f}")
```

### Cell 5: Word Arithmetic (The Famous Example!)
```python
# Cell 5: Vector math - king - man + woman = ?
print("=" * 50)
print("🧮 WORD ARITHMETIC")
print("=" * 50)

# Classic example: king - man + woman
result = model.most_similar(positive=['king', 'woman'], negative=['man'], topn=5)
print("\n👑 'king' - 'man' + 'woman' = ")
for word, score in result:
    print(f"   {word}: {score:.3f}")

# More examples
print("\n🌍 'paris' - 'france' + 'germany' = ")
result = model.most_similar(positive=['paris', 'germany'], negative=['france'], topn=3)
for word, score in result:
    print(f"   {word}: {score:.3f}")

print("\n👨 'doctor' - 'man' + 'woman' = ")
result = model.most_similar(positive=['doctor', 'woman'], negative=['man'], topn=3)
for word, score in result:
    print(f"   {word}: {score:.3f}")
```

### Cell 6: Calculate Similarity Scores
```python
# Cell 6: Measure how similar different word pairs are
print("=" * 50)
print("📏 SIMILARITY SCORES")
print("=" * 50)

pairs = [
    ('king', 'queen'),
    ('apple', 'orange'),
    ('car', 'bicycle'),
    ('king', 'apple'),
    ('happy', 'joyful'),
    ('hot', 'cold'),
    ('computer', 'banana')
]

print("\n🔢 Cosine Similarity (1 = identical, 0 = unrelated, -1 = opposite):\n")
for word1, word2 in pairs:
    similarity = model.similarity(word1, word2)
    print(f"   {word1:10} ↔ {word2:10}: {similarity:.3f}")
```

### Cell 7: Find Odd One Out
```python
# Cell 7: Which word doesn't belong?
print("=" * 50)
print("🎲 FIND THE ODD ONE OUT")
print("=" * 50)

word_sets = [
    ['king', 'queen', 'prince', 'apple'],
    ['car', 'bicycle', 'train', 'banana'],
    ['happy', 'joyful', 'excited', 'table'],
    ['python', 'java', 'javascript', 'elephant']
]

for words in word_sets:
    odd_one = model.doesnt_match(words)
    print(f"\n📝 Words: {words}")
    print(f"   ❌ Odd one out: {odd_one}")
```

### Cell 8: Compare Tokenization vs Embeddings
```python
# Cell 8: Understanding why embeddings matter
print("=" * 50)
print("🆚 TOKENIZATION vs EMBEDDINGS")
print("=" * 50)

# Tokenization approach (just IDs)
tokens = {
    'king': 5234,
    'queen': 8923,
    'man': 3451,
    'woman': 6782
}

print("\n🔢 Tokenization (just IDs):")
print(f"   king → {tokens['king']}")
print(f"   queen → {tokens['queen']}")
print(f"   Difference: {abs(tokens['king'] - tokens['queen'])} (meaningless!)")

# Embedding approach
print("\n📊 Embeddings (meaning vectors):")
king_vec = model['king']
queen_vec = model['queen']
similarity = model.similarity('king', 'queen')
print(f"   Similarity between 'king' and 'queen': {similarity:.3f}")
print(f"   ✅ This captures meaning! High similarity = related concepts")

# Contrast with unrelated words
apple_similarity = model.similarity('king', 'apple')
print(f"\n   Similarity between 'king' and 'apple': {apple_similarity:.3f}")
print(f"   ✅ Low similarity = unrelated concepts")
```

### Cell 9: Visualize Embeddings in 2D
```python
# Cell 9: Create a visualization of word embeddings
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

print("=" * 50)
print("🎨 VISUALIZING EMBEDDINGS")
print("=" * 50)

# Select words to visualize
words = ['king', 'queen', 'prince', 'princess',
         'man', 'woman', 'boy', 'girl',
         'apple', 'orange', 'banana', 'grape',
         'car', 'truck', 'bicycle', 'bus']

# Get their vectors
vectors = np.array([model[word] for word in words])

print(f"📊 Visualizing {len(words)} words...")

# Reduce from 50D to 2D
tsne = TSNE(n_components=2, random_state=42, perplexity=5)
vectors_2d = tsne.fit_transform(vectors)

# Create the plot
plt.figure(figsize=(14, 10))

# Color code by category
colors = []
for word in words:
    if word in ['king', 'queen', 'prince', 'princess']:
        colors.append('red')  # Royalty
    elif word in ['man', 'woman', 'boy', 'girl']:
        colors.append('blue')  # People
    elif word in ['apple', 'orange', 'banana', 'grape']:
        colors.append('green')  # Fruits
    else:
        colors.append('purple')  # Vehicles

# Plot each word
for i, word in enumerate(words):
    x, y = vectors_2d[i]
    plt.scatter(x, y, c=colors[i], s=100)
    plt.annotate(word, (x, y), fontsize=12, fontweight='bold')

plt.title("Word Embeddings Visualized in 2D Space", fontsize=16)
plt.xlabel("Dimension 1", fontsize=12)
plt.ylabel("Dimension 2", fontsize=12)
plt.grid(True, alpha=0.3)

# Add legend
from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor='red', label='Royalty'),
    Patch(facecolor='blue', label='People'),
    Patch(facecolor='green', label='Fruits'),
    Patch(facecolor='purple', label='Vehicles')
]
plt.legend(handles=legend_elements, loc='best')

plt.tight_layout()
plt.show()

print("\n💡 OBSERVATIONS:")
print("   • Similar words cluster together in space")
print("   • Royalty words (red) are close to each other")
print("   • Fruits (green) form their own cluster")
print("   • The distance between words represents semantic similarity")
```

### Cell 10: Practical Application - Semantic Search
```python
# Cell 10: Build a simple semantic search engine
print("=" * 50)
print("🔍 SEMANTIC SEARCH DEMO")
print("=" * 50)

# Create a knowledge base
knowledge_base = {
    "Tokenization": "Breaking text into smaller units called tokens",
    "BPE": "Byte Pair Encoding merges frequent character pairs",
    "WordPiece": "Tokenizer used by BERT that uses probability scores",
    "Embeddings": "Converting tokens to dense vectors capturing meaning",
    "Transformers": "Neural network architecture using attention",
    "Attention": "Mechanism that weighs importance of different tokens"
}

# Embed all documents
documents = list(knowledge_base.keys())
doc_vectors = np.array([model.get_vector(doc.lower()) for doc in documents])

def semantic_search(query, top_k=2):
    """Find most relevant documents for a query"""
    # Embed the query
    query_vector = model.get_vector(query.lower())
    
    # Calculate similarity with all documents
    similarities = []
    for i, doc_vec in enumerate(doc_vectors):
        # Cosine similarity
        sim = np.dot(query_vector, doc_vec) / (np.linalg.norm(query_vector) * np.linalg.norm(doc_vec))
        similarities.append(sim)
    
    # Get top matches
    top_indices = np.argsort(similarities)[::-1][:top_k]
    
    return [(documents[i], similarities[i]) for i in top_indices]

# Test the search
test_queries = [
    "How to split text into pieces?",
    "What gives words their meaning?",
    "Neural networks for language"
]

for query in test_queries:
    print(f"\n📝 Query: '{query}'")
    results = semantic_search(query)
    print(f"   Top match: {results[0][0]} (score: {results[0][1]:.3f})")
    print(f"   Description: {knowledge_base[results[0][0]]}")
```
