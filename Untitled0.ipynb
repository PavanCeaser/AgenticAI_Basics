```python
# Install and import
!pip install nbformat requests gdown

import nbformat
import requests
import json
import io
import gdown

# Method 1: If you have the notebook saved in Google Drive
# Get the file ID from the shareable link
# Go to your notebook in Colab → Share → Copy link
# It looks like: https://colab.research.google.com/drive/1xOCohASvyCza-Gx5ILGJTzx-eq1tDcXp

file_id = "1xOCohASvyCza-Gx5ILGJTzx-eq1tDcXp"  # Just the ID part

# Download using gdown (handles Google Drive properly)
output = f'/content/{file_id}.ipynb'
gdown.download(f'https://drive.google.com/uc?id={file_id}', output, quiet=False)

# Load and clean
with open(output, 'r') as f:
    notebook = nbformat.read(f, as_version=4)

# Remove widgets metadata from all cells
for cell in notebook.cells:
    if 'metadata' in cell and 'widgets' in cell['metadata']:
        del cell['metadata']['widgets']

# Remove from notebook metadata
if 'metadata' in notebook and 'widgets' in notebook['metadata']:
    del notebook['metadata']['widgets']

# Save the cleaned notebook
with open('cleaned_notebook.ipynb', 'w') as f:
    nbformat.write(notebook, f)

print("✅ Notebook cleaned successfully!")
```
